#**Alta Cardinalidade:**

```python
#transformando o dicionario em um dataframe
dicionario = pd.DataFrame.from_dict(dicionario)
```

```python
#selecionando apenas as colunas nominais
colunas_nominais = dicionario.query('type=="nominal"').variavel.values
colunas_nominais
```

```python
#retornando a cardinalidade de cada coluna
for coluna in colunas_nominais:

  x = len(df[coluna].unique())
  y = len(df[coluna])

  c = x/y

  # Limiar da alta cardinalidade
  threshold = 0.5

  # Conferindo se a coluna possui alta cardinalidade
  if c > threshold:
    print(f'{coluna}: APRESENTA ALTA CARDINALIDADE')
  else:
    print(f'{coluna}: NÃO APRESENTA ALTA CARDINALIDADE')
```

Como observado as colunas categóricas analisadas não possuem alta cardinalidade, o que significa que podemos ter mais clareza e eficiencia na analise de dados.

#**Colunas Desbalanceadas**

🔸Se a coluna for balancada significa que possui uma distribuição balanceada
de valores entre suas categorias.

🔸Se a coluna for desbalanceada siginifica que possui distribuição desbalanceada entre suas categorias.

```python
#Laço repetitivo para passar por todas as colunas
for colunas in colunas_nominais:

  min = df[colunas].value_counts().min()

  # Calculando a frequência mínima
  max = df[colunas].value_counts().max()

  # Calculando o desbalanceamento
  d = min / max

  # Limiar do desbalanceamento
  threshold_d = 0.5

  # Conferindo o balaceamento da coluna
  if d < threshold_d:
    print(f'{colunas}: DESBALANCEADA')
  else:
    print(f'{colunas}: BALANCEADA')
```

Possuem 7 colunas balanceadas e 6 colunas desbalanceadas.

Nas colunas *nacionalidade* e *PlaceofBirth* podemos agrupar por paises que estão mais próximos um do outro.

Na coluna *Topic* podemos agrupar por tipo de matéria como exatas, humanas e linguagens.

Para as colunas *stageID* e *GradeID* pode se considerar a coleta de mais dados.

#**Valores zeros:**

```python
#Retornando as colunas discretas do dataframe
colunas_discreta = dicionario.query('type=="discreta"').variavel.values
colunas_discreta
```

```python
# Calculando o total de valores zeros dentro de um estrutura de repetição

for coluna in colunas_discreta:

  total_values = len(df[coluna])

  # Calculando os valores zeros
  zero_count = (df[coluna ]== 0).sum()

  # Calculando proporção dos valores zeros
  zero_ratio = zero_count / total_values

  # Identificando se apresenta muitos valores zeros
  threshold_zero = 0.5

  if  zero_ratio > threshold_zero:
    print('APRESENTA MUITOS VALORES 0')
  else:
    print('NÃO APRESENTA MUITOS VALORES 0')
```
